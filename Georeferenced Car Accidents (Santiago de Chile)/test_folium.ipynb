```python
import folium

# Crear el objeto de mapa en la localizaci√≥n inicial con un zoom determinado
m = folium.Map(location=[45.5236, -122.6750], zoom_start=13)

# Agregar un marcador al mapa
folium.Marker(
    location=[-33.364435, -70.798508],
    popup='The Waterfront',
    icon=folium.Icon(icon='cloud'),
).add_to(m)

# Mostrar el mapa
m
```
